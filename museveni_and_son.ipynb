<h2> Loading an Image in Python</h2>

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
#%matplotlib inline  # if you are running this code in Jupyter notebook

# reads image 'opencv-logo.png' as grayscale
img = cv2.imread('FRFW7d_XwAAoG8a.jpg', cv2.IMREAD_GRAYSCALE) # the source of the image is
# https://pbs.twimg.com/media/FRFW7d_XwAAoG8a?format=jpg&name=medium
plt.imshow(img, cmap='gray')
```

<h2> Plotting the Image</h2>

```python
plt.figure(figsize=(10,10))
plt.imshow(img)
plt.show()
```

```python
new_image=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(new_image)
plt.show()
```

```python
image_gray = cv2.cvtColor(new_image, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(image_gray, cmap='gray')
plt.show()
```
